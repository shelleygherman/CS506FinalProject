```python
file = pd.read_csv('ICPSR_36404/DS0001/36404-0001-Data.tsv', sep='\t')
```

```python
df = pd.DataFrame(file)
```

```python
df55Plus = df.loc[df['AGEADMIT']==5]
```

```python
dfFemales = df.loc[df['SEX']==2]
dfMales = df.loc[df['SEX']==1]
```

```python
len(dfFemales)
```

```python
len(dfMales)
```

```python
df1 = pd.DataFrame({'countFemales' : dfFemales.groupby( ["STATE"] ).size()}).reset_index()
df1 = df1.replace({'STATE' : states})
df1.index = df1.STATE

df2 = pd.DataFrame({'countMales' : dfMales.groupby( ["STATE"] ).size()}).reset_index()
df2 = df2.replace({'STATE' : states})
df2.index = df2.STATE

df3 = df1.merge(df2, how = 'inner', on = 'STATE')
```

```python
df3.index = df3.STATE
```

```python
df3.plot.bar(title = 'Men vs Women over 55 by State in Prison', figsize=(15, 9))
```

```python
dfWhite = df.loc[df['RACE']==1]
dfBlack = df.loc[df['RACE']==2]
dfHispanic = df.loc[df['RACE']==3]
dfOther = df.loc[df['RACE']==4]
dfMissing = df.loc[df['RACE']==9]
```

```python
R = [len(dfWhite), len(dfBlack), len(dfHispanic), len(dfOther), len(dfMissing)]
labels = ['White', 'Black', 'Hispanic', 'Other', 'Missing']
pd.Series(index=labels, data=R).plot(kind = 'bar', title= '55+ Prisoners by Race', figsize = (10, 6))
```
